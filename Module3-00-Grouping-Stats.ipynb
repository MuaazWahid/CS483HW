**Step7** Group the data below using the mapping and find the sum\
for each person across the columns, for example, a, b, and e are `red`\
c and e are `blue`, so Joe will have the sum of the columns correspoding\
to `red`columns as well as `blue` color.

```python
df = pandas.DataFrame({'Joe':[1,2,3,4,5],
                   'Steve':[6,7,8,9,10],
                   'Alice':[11,12,13,14,15]},
                  index=['a','b','c','d','e'])

mapping = {'a':'red', 'b':'red', 'c':'blue', 'd':'red', 'e':'blue'}

# Create a new column with the color mapping
df['color'] = [mapping[i] for i in df.index]


# Group by color and sum across columns
result = df.groupby('color').sum()
print(result)
```

```python
stock.rename(columns = {'Unnamed: 0': 'date'}, inplace=True)
stock.head()
```

```python
stock['date'] = pandas.to_datetime(stock['date'], format = '%Y-%m-%d %H:%M:%S')
stock.head()
```

```python
stock['year'] = stock['date'].dt.year
stock.head()
```

```python
# index, set_index, index_reset,
stock.set_index('date', inplace=True)
stock.head()
```

```python
stock.reset_index()
```

```python
stock[(stock.index.year==2003) & (stock.index.month==1)]
```

```python
def spx_correlation(group):
    return group.corrwith(group['SPX'])
```

```python
#corr() finds the pairwise correlation between columns of a dataframe
#corrwith() finds correlation between dataframes or correlation betweena. dataframe and a series
```

```python
stock.head()
```

```python
st_year = stock.groupby('year')[['AAPL','MSFT','XOM','SPX']]
st_year.mean()
```

```python
st_year.apply(spx_correlation)
```
