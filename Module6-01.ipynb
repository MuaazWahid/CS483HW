## 1. Install the libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 2. Loading a Dataset
We'll use the `tips` dataset. This dataset contains information about tips received by waitstaff at a restaurant, including the total bill, the size of the party, and the gender of the person who paid the bill.

```python
df = pd.read_csv('tips.csv')
df.head(5)
```

## 3. Scatter Plot: Total Bill vs Tip
A scatter plot is useful for visualizing the relationship between the total bill and the tip.

```python
# Scatter plot for total_bill vs tip
plt.scatter(df['total_bill'], df['tip'], alpha = 0.5, color='green')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.grid()
plt.show()
```

## 4. Line Plot: Average Tip Over Days
You can use a line plot to visualize trends, such as the average tip on each day.

```python
# Line plot showing average tip per day
avg_tip_by_day = df.groupby('day')['tip'].mean()

# plot
plt.plot(avg_tip_by_day.index, avg_tip_by_day.values, marker='o')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip')
plt.title('Average Tip over Days')
plt.grid()
plt.show()
```

## 5. Box Plot: Distribution of Tips by Day
A box plot is useful to show the distribution of tips by day.

```python
# Box plot for tips distribution by day
tip_data_by_day = [df['tip'][df['day'] == day] for day in df['day'].unique()]

plt.boxplot(tip_data_by_day, labels=df['day'].unique())
plt.xlabel('Day')
plt.ylabel('Tip')
plt.title('Box Plot of Tips by Day')
plt.show()
```

## 6. Pie Chart: Distribution of Smokers vs Non-Smokers
A pie chart shows proportions. We'll visualize the proportion of smokers vs non-smokers in the dataset.

```python
# Pie chart showing smoker vs non-smoker distribution
smoker_counts = df['smoker'].value_counts()
plt.pie(smoker_counts.values, labels = smoker_counts.index, colors = ['lightskyblue', 'lightcoral'],
autopct = '%1.1f%%', startangle = 90)
plt.title('Smoker vs Non-Smoker Proportion')
plt.show()
```

## 7. Time Series: Number of Tips Over Time (Simulated)
Since the tips dataset doesnt contain a time series column, let's simulate a time series by creating an artificial time column.

```python
# Simulate a time series (create a date range)
#dates = pd.date_range(start='2023-01', end='2023-09', freq='D')
dates = pd.date_range(start='2023-01', periods=len(df), freq='D')
df['date'] = dates
daily_total_bill = df.groupby('date')['total_bill'].sum()

# Plot the time series
plt.plot(daily_total_bill.index, daily_total_bill.values)
plt.title('Time Series of Total Bill Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.show()
```

## 8. Bar Graph: Average Total Bill by Day
A bar graph is helpful to compare categorical data, such as the average total bill per day.

```python
# Bar graph for average total bill by day
avg_bill_by_day = df.groupby('day')['total_bill'].mean()

# Create bar graph
plt.bar(avg_bill_by_day.index, avg_bill_by_day.values)
plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.title('Average Total Bill by Day')
plt.show()
```
