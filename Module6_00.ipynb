## 1. Import the necessary libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 2. Scatter Plot

A scatter plot is used to visualize the relationship between two numerical variables.



```python
# Sample data
x1 = np.random.rand(50)
x2 = np.random.rand(50)
y2 = np.random.randint(0,2,50)

# Scatter plot
plt.scatter(x1,x2,  alpha=0.6,c=y2, cmap = 'bwr',label='Data Points')
plt.title('Scatter Plot')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
```

## 3. Line Plot
A line plot is often used to display trends over time.

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Line plot
plt.plot(x, y)
plt.title('Line Plot of sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

## 4. Box Plot
A box plot (or box-and-whisker plot) is used to display the distribution of data and identify outliers.

```python
# Sample data
data = np.random.randn(100)

# Box plot
plt.boxplot(data)
plt.title('Box Plot')
plt.show()
```

## 5. Pie Chart
A pie chart represents data as proportions of a whole.

```python
# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [20, 30, 25, 25]

# Pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

## 6. Time Series Plot
A time series plot is used to visualize data over time.

```python
# Generate a time series dataset
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
data = np.random.randn(100).cumsum()

# Time series plot
plt.plot(dates, data)
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.show()
```

## 7. Bar Graph
A bar graph is used to compare values across categories.

```python
# Sample data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# Bar graph
plt.bar(categories, values)
plt.title('Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

## 8. Combining Multiple Plots in One Figure
You can also combine multiple plots into one figure using plt.subplot().

```python
# Set up the subplot grid
plt.figure(figsize=(10, 8))

# Scatter plot
plt.subplot(2,2,1)
x1 = np.random.rand(50)
x2= np.random.rand(50)
y2 = np.random.randint(0,2,50)
plt.scatter(x1, x2, c=y2, cmap = 'bwr', label='Data Points', marker='x')
plt.title('Scatter Plot')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc= 'upper left')

# Line plot
plt.subplot(2,2,2)
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y, color='r')

# Box plot
plt.subplot(2,2,3)
data1= np.random.normal(0, 1, 1000)
data2= np.random.normal(20,10,1000)
data3= np.random.normal(10,4,1000)
plt.boxplot( [data1, data2, data3])

# Bar graph
plt.subplot(2,2,4)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
plt.bar(categories, values)

# Show all plots
plt.tight_layout()
```

## 9. Adding Customizations
You can add several customizations to your plots, such as labels, gridlines, legends, and changing color schemes. Here’s how you can add some of these features:

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, marker='x', color='red', label='Points')
plt.title('Customized Scatter Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.grid()
plt.legend()
plt.show()
```
